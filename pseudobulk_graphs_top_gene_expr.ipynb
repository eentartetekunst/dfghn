```python
single = '/Users/sofialuk/Desktop/thesis/h1299/h1299_desingle/desingle_h1299_36h.csv' # desingle DEGs
pseudo = '/Users/sofialuk/Desktop/thesis/h1299/h1299_deseq/h1299_DEGs_36h.csv' # deseq DEGs
expr = '/Users/sofialuk/Desktop/thesis/h1299/h1299_counts/h1299_36hpi.csv' # count matrix 

desingle = pd.read_csv(single)
deseq = pd.read_csv(pseudo)
counts = pd.read_csv(expr)
```

### top genes by lfc in desingle

remove inf fold change and log2

```python
desingle.replace([np.inf, -np.inf], np.nan, inplace=True)
desingle.dropna(subset = ['norm_foldChange'], inplace=True)

# desingle = desingle.sort_values(by='norm_foldChange', ascending=False, key=abs).iloc[:, :]
desingle = desingle[desingle['pvalue.adj.FDR'] < 0.01]
desingle['log2FoldChange'] = np.log2(desingle['norm_foldChange'])
desingle = desingle.sort_values(by='log2FoldChange', ascending=False, key=abs).iloc[:, :]

desingle.iloc[:, 10:]
```

```python
top_single = desingle.head(20).index
top_single
```

### top genes by lfc in deseq

```python
deseq = deseq[deseq['padj'] < 0.01]
deseq = deseq.sort_values(by='log2FoldChange', ascending=False, key=abs)
```

```python
deseq.index = deseq['Unnamed: 0']
deseq = deseq.drop('Unnamed: 0', axis=1)
deseq.head()
```

```python
top_pseudo = deseq.head(20).index
top_pseudo
```

```python
#counts.index = counts['Unnamed: 0']
#counts = counts.drop('Unnamed: 0', axis=1)
counts.head()
```

```python
def prep_counts(counts, top):
    expr = counts[counts['Unnamed: 0'].isin(top)].copy()
    matrix = expr.melt('Unnamed: 0')
    matrix['var'] = matrix['variable'].str.slice(6, 8) # s2 / mock=mo
    
    return matrix 
```

```python
pseudo = prep_counts(counts, top_pseudo)
single = prep_counts(counts, top_single)
```
